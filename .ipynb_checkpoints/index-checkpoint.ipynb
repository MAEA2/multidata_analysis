# Bスプライン関数での回帰

## Bスプライン関数の描画
$0$次のB-spline関数については
$$
b_j(x;0)=\begin{cases}
1 & x\in [t_j,t_{j+1}]\\
0 & x \notin [t_j,t_{j+1}]
\end{cases}
,\quad j=1,2,\ldots,m
$$
で定められる。$r\geq 1$次については以下の漸化式により求める。

$$
b_j(x;r)=\frac{x-t_j}{t_{j+r}-t_i}b_{j}(x;r-1)+\frac{t_{j+r+1}-x}{t_{j+r+1}-t_{j+1}}b_{j+1}(x;r-1)
$$


```python
%matplotlib inline  
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def b(x,j,r,h):
    if r==0:
        if x >= j*h and x < (j+1)*h:
            return 1
        else:
            return 0
    else:
        return (x-j*h)/((j+r)*h-j*h)*b(x,j,r-1,h)+((j+r+1)*h-x)/((j+r+1)*h-(j+1)*h)*b(x,j+1,r-1,h)

x=np.arange(0,10,0.01)

for r in range(6):
    for j in range(-2,10):
        y=[b(x_val,j,r,1) for x_val in x]
        plt.hold(True)
        plt.plot(x,y,label="b(x,%d,%d)" % (j,r))
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
    plt.ylim(0,1.1)
    plt.show()
```

## B-spline関数による回帰

YAHOOの株価のデータをpandasのdatareaderを用いて取り出し、3次B-spline関数による回帰を行なってみる。

```python
import pandas_datareader.data as web
import datetime

start = datetime.datetime(2013, 1, 1)
end = datetime.datetime(2017, 5, 5)
d=web.DataReader("YHOO","yahoo",start,end)#訓練用データ
n_data=len(d) #全データ数
n=1000 #訓練データ数

xlist=np.arange(0,n,1)
ylist=list(d["Adj Close"][:n]) #Adj Closeは調整後の株価の終値：目的変数

x_test=np.arange(n,n_data,1)
y_test=list(d["Adj Close"][n:n_data])

def Base(x,m):
    BB=list(b(x,j,3,h) for j in range(-2,m))
    return np.append(1,BB)

for m in [10,20,30]:
    h=n/m
    B = np.array([Base(x,m) for x in xlist])
    w = np.linalg.solve(np.dot(B.T, B), np.dot(B.T,ylist))

    x_ac=np.arange(0,n_data,1)

    y_est = [np.dot(w, Base(x,m)) for x in x_ac]

    for j in range(-2,m):
        y=[20*b(x,j,r,h) for x in x_ac]
        plt.plot(x_ac,y,"y-")
    plt.plot(xlist,ylist, 'b.')
    plt.plot(x_ac,y_est,"r-")
    plt.plot(x_test,y_test,"g.")
    plt.xlim(0,n_data)
    plt.xlabel("day:from %s to %s" % (start,end))
    plt.ylabel("Adjusted Closing Price")
    plt.title("number of basis=%d" % (m+3))
    plt.show()
```

## B-spline回帰に対する結果と考察
定数関数と合わせて$13,23,33$個で回帰を行なった。回帰はそこそこに行えていることがわかる。
予測に関しては見事に失敗している。

# 回帰分析の実験

ボストンの住宅価格データhttps://archive.ics.uci.edu/ml/datasets/Housing
に対して重回帰を行った。

```python
import pandas as pd
import numpy as np
from sklearn import linear_model as lm

index=pd.DataFrame(["犯罪率","宅地の割合","非商用地の割合","チャールズ流域かどうか",
                    "窒素酸化物濃度","平均部屋数","築年数",
                    "ビジネス地区への距離","高速道路へのアクセス指数",
                    "固定資産税","学生と教師の割合","黒人の割合","低所得者の割合"],
      index=["crim","zn","indus","chas","nox","rm","age","dis","rad","tax","ptratio","black","lstat"],
      columns=["説明変数の意味"]
           )
index
```

```python
df = pd.read_csv("Boston.csv", index_col=0)

df
```

medvが住宅価格の中央値である。これを目的変数とする。

```python
y_tr = df.iloc[:400,  13].values #目的変数
y_test = df.iloc[400:,13].values

df = (df - df.mean())/df.std() #基準化

X_tr = df.iloc[:400, :13].values #説明変数の行列
X_test = df.iloc[400:,:13].values 


linear=lm.LinearRegression()
lasso = lm.Lasso(alpha=0.1)
ridge = lm.Ridge(alpha=0.1)

linear.fit(X_tr,y_tr)
ridge.fit(X_tr,y_tr)
lasso.fit(X_tr,y_tr)

result_coef = pd.DataFrame([linear.coef_,ridge.coef_,lasso.coef_],index = ["Linear","Ridge","Lasso"], columns=list(df.columns.values)[0:13])
result_coef
```

# 推定結果についての考察

$\alpha$は平滑化パラメータである。
Lasso推定においては確かにいくつかの係数が$0$に落ちていることが確認できた。

# 平滑化パラメータへの依存性

$\alpha$に対する依存性をみる。

```python

def plot_result(alpha):
    linear=lm.LinearRegression()
    lasso = lm.Lasso(alpha=alpha)
    ridge = lm.Ridge(alpha=alpha)
    
    linear.fit(X_tr,y_tr)
    ridge.fit(X_tr,y_tr)
    lasso.fit(X_tr,y_tr)
    
    result_coef = pd.DataFrame([linear.coef_,ridge.coef_,lasso.coef_],index = ["Linear","Ridge","Lasso"], 
                               columns=list(df.columns.values)[0:13])
    result_coef.plot.bar()
    plt.hold(True)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
    plt.title("alpha=%f " % alpha)
    
plot_result(0.001)
plot_result(0.1)
plot_result(0.5)
plot_result(1.0)
```

## テストデータを用いた実験
lassoについては$\alpha$によって係数が鋭敏に変化していることがわかった。そこで訓練データより推定した係数$\boldsymbol{\beta}$を用いて、テストデータをどのていど予測できているかをみる。その際、$\alpha$を変化させたときテストデータについての平均誤差にどのように影響を及ぼすかをみる。


```python
alpha=np.arange(10E-10,2.0,0.01)

lin_E=[]
ridge_E=[]
lasso_E=[]

for a in alpha:
    linear=lm.LinearRegression()
    ridge = lm.Ridge(alpha=a)
    lasso = lm.Lasso(alpha=a)

    linear.fit(X_tr,y_tr)
    lasso.fit(X_tr,y_tr)
    ridge.fit(X_tr,y_tr)

    lin_error =0
    ridge_error=0
    lasso_error =0
    n_test=len(y_test)
    for i in range(n_test):
        lin_error += (y_test[i]-sum(X_test[i,]*linear.coef_)-linear.intercept_)**2/n_test
        ridge_error+=(y_test[i]-sum(X_test[i,]*ridge.coef_)-ridge.intercept_)**2/n_test
        lasso_error +=(y_test[i]-sum(X_test[i,]*lasso.coef_)-lasso.intercept_)**2/n_test
    lin_E.append(lin_error)
    ridge_E.append(ridge_error)
    lasso_E.append(lasso_error)

plt.hold(True);
plt.xlabel("alpha")
plt.ylabel("Mean Squared Error")
plt.plot(alpha,lin_E,label="linear") 
plt.plot(alpha,ridge_E,label="ridge")
plt.plot(alpha,lasso_E,label="lasso")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.show()
```
