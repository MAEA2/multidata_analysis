# データ解析例

ボストンの住宅価格データhttps://archive.ics.uci.edu/ml/datasets/Housing
に対して重回帰を行った。

```python
import pandas as pd
import numpy as np
from sklearn import linear_model as lm

index=pd.DataFrame(["犯罪率","宅地の割合","非商用地の割合","チャールズ流域かどうか",
                    "窒素酸化物濃度","平均部屋数","築年数",
                    "ビジネス地区への距離","高速道路へのアクセス指数",
                    "固定資産税","学生と教師の割合","黒人の割合","低所得者の割合"],
      index=["crim","zn","indus","chas","nox","rm","age","dis","rad","tax","ptratio","black","lstat"],
      columns=["説明変数の意味"]
           )
index
```

```python
df = pd.read_csv("Boston.csv", index_col=0)

df
```

medvが住宅価格の中央値である。これを目的変数とする。

```python
y_tr = df.iloc[:400,  13].values #目的変数
y_test = df.iloc[400:,13].values

df = (df - df.mean())/df.std() #基準化

X_tr = df.iloc[:400, :13].values #説明変数の行列
X_test = df.iloc[400:,:13].values 


linear=lm.LinearRegression()
lasso = lm.Lasso(alpha=0.1)
ridge = lm.Ridge(alpha=0.1)

linear.fit(X_tr,y_tr)
ridge.fit(X_tr,y_tr)
lasso.fit(X_tr,y_tr)

result_coef = pd.DataFrame([linear.coef_,ridge.coef_,lasso.coef_],index = ["Linear","Ridge","Lasso"], columns=list(df.columns.values)[0:13])
result_coef
```

# 推定結果についての考察

$alpha$は平滑化パラメータである。
Lasso推定においては確かにいくつかの係数が0に落ちていることが確認できた。

# 平滑化パラメータへの依存性

$\alpha$に対する依存性をみる。

```python
from ipywidgets import interact
import matplotlib.pyplot as plt
import pylab

def plot_result(alpha):
    linear=lm.LinearRegression()
    lasso = lm.Lasso(alpha=alpha)
    ridge = lm.Ridge(alpha=alpha)
    
    linear.fit(X_tr,y_tr)
    ridge.fit(X_tr,y_tr)
    lasso.fit(X_tr,y_tr)
    
    result_coef = pd.DataFrame([linear.coef_,ridge.coef_,lasso.coef_],index = ["Linear","Ridge","Lasso"], 
                               columns=list(df.columns.values)[0:13])
    result_coef.plot.bar()
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
    plt.title("alpha=%f " % alpha)
    
plot_result(0.001)
plot_result(0.1)
plot_result(0.5)
plot_result(1.0)
    
```

## テストデータを用いた実験
lassoについては$\alpha$によって係数が鋭敏に変化していることがわかった。そこで訓練データより推定した係数$\boldsymbol{\beta}$を用いて、テストデータをどのていど予測できているかをみる。その際、$\alpha$を変化させたときテストデータについての平均誤差にどのように影響を及ぼすかをみる。


```python
alpha=np.arange(10E-10,2.0,0.01)

lin_E=[]
ridge_E=[]
lasso_E=[]

for a in alpha:
    linear=lm.LinearRegression()
    ridge = lm.Ridge(alpha=a)
    lasso = lm.Lasso(alpha=a)

    linear.fit(X_tr,y_tr)
    lasso.fit(X_tr,y_tr)
    ridge.fit(X_tr,y_tr)

    lin_error =0
    ridge_error=0
    lasso_error =0
    n_test=len(y_test)
    for i in range(n_test):
        lin_error += (y_test[i]-sum(X_test[i,]*linear.coef_)-linear.intercept_)**2/n_test
        ridge_error+=(y_test[i]-sum(X_test[i,]*ridge.coef_)-ridge.intercept_)**2/n_test
        lasso_error +=(y_test[i]-sum(X_test[i,]*lasso.coef_)-lasso.intercept_)**2/n_test
    lin_E.append(lin_error)
    ridge_E.append(ridge_error)
    lasso_E.append(lasso_error)

plt.hold(True);
plt.xlabel("alpha")
plt.ylabel("Mean Squared Error")
plt.plot(alpha,lin_E,label="linear") 
plt.plot(alpha,ridge_E,label="ridge")
plt.plot(alpha,lasso_E,label="lasso")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)


```
