```python
%matplotlib inline

# グラフ化に必要なものの準備
import matplotlib
import matplotlib.pyplot as plt
from ipywidgets import interact

# matplotlibのデフォルトフォントをTakaoGothicに設定
font = {'family' : 'TakaoGothic'}
matplotlib.rc('font', **font)

def plot_coef(alpha):
    linear=lm.LinearRegression()
    lasso = lm.Lasso(alpha=alpha)
    ridge = lm.Ridge(alpha=alpha)
    linear.fit(X_tr,y_tr)
    ridge.fit(X_tr,y_tr)
    lasso.fit(X_tr,y_tr)
    result_coef = pd.DataFrame([linear.coef_,ridge.coef_,lasso.coef_],index = ["Linear","Ridge","Lasso"], 
                               columns=list(df.columns.values)[0:13])
    lin_error =0
    ridge_error=0
    lasso_error =0
    n_test=len(y_test)
    for i in range(n_test):
        lin_error += (y_test[i]-sum(X_test[i,]*linear.coef_)-linear.intercept_)**2/n_test
        ridge_error+=(y_test[i]-sum(X_test[i,]*ridge.coef_)-ridge.intercept_)**2/n_test
        lasso_error +=(y_test[i]-sum(X_test[i,]*lasso.coef_)-lasso.intercept_)**2/n_test
    result_error = pd.DataFrame([lin_error,ridge_error,lasso_error],index = ["Linear","Ridge","Lasso"],columns=["Average Square Eroor"])
    result_coef.plot.bar()
    plt.title("parameters", size=16)
    plt.legend(loc="upper center", 
bbox_to_anchor=(1.3,1.0), 
ncol=2)
    result_error.plot.bar()
    plt.legend(loc="upper center", 
    bbox_to_anchor=(1.3,1.0), 
    ncol=2)
    plt.title("Average Sauare Error", size=16)


interact(plot_coef,alpha=(1E-10,1.0,0.01))
```
